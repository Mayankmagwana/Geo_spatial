```python
import geopandas as gpd
import pandas as pd
import folium
```

```python
from folium.plugins import HeatMap, MarkerCluster
from folium import Choropleth, Circle, Marker
```

```python
m_1 = folium.Map(location=[42.31,-71.0589], tiles = "openstreetmap", zoom_start=10)
m_1
```

```python
crime_df = pd.read_csv("Kaggle/crimes-in-boston/crimes-in-boston/crime.csv")
```

```python
crime_df.head()
```

```python
crime_df = crime_df[crime_df.OFFENSE_CODE_GROUP.isin(['Larceny', 'Auto Theft', 'Robbery', 'Larceny From Motor Vehicle', 'Residential Burglary',
    'Simple Assault', 'Harassment', 'Ballistics', 'Aggravated Assault', 'Other Burglary', 
    'Arson', 'Commercial Burglary', 'HOME INVASION', 'Homicide', 'Criminal Harassment', 
    'Manslaughter'])]
```

```python
crime_df.dropna(subset = ["Lat","Long","DISTRICT"],inplace = True)
crime_df = crime_df[crime_df["YEAR"]>=2018]
crime_df = crime_df[crime_df["OFFENSE_CODE_GROUP"] == "Robbery"][crime_df["HOUR"].isin(range(9,18))]
```

```python
crime_df.head()
```

```python
crime = gpd.GeoDataFrame(crime_df, geometry = gpd.points_from_xy(crime_df.Lat,crime_df.Long))
```

```python
crime.crs  = {"init":"epsg:4326"}
crime.head()
```

```python
boston = folium.Map(location = [42.31,-71.0589], titles = "openstreetmap", zoom_start = 12)

for indx, row in crime.iterrows():
    Marker([row['Lat'],row['Long']]).add_to(boston)
boston
```

Cluster Problem


```python
import math
```

```python
boston = folium.Map(location=[42.32,-71.0589],titles= "cartodbpositron",zoom_start= 13)
mc = MarkerCluster()
for indx, row in crime.iterrows():
    if not math.isnan(row['Long']) and not math.isnan(row["Lat"]):
        mc.add_child(Marker([row['Lat'],row['Long']]))
boston.add_child(mc)
boston
```

Bubble Map

```python
boston = folium.Map(location=[42.32,-71.0589],titles='cartodbpositron',zoom_start = 13)
def color(val):
    if val <= 12:
        return 'forestgreen'
    else:
        return 'darkred'

    
for i in range(0,len(crime)):
    Circle(location=[crime.iloc[i]["Lat"],crime.iloc[i]["Long"]],
           radius =20,color=color(crime.iloc[i]["HOUR"])).add_to(boston)
boston
```

Heatmaps

```python
boston = folium.Map(location=[42.32,-71.0589],titles="openstreetmap",zoom_start=12)

HeatMap(data=crime[['Lat','Long']], radius = 10).add_to(boston)
boston
```

```python
boundries = gpd.read_file("Kaggle/Plate_Boundaries/Plate_Boundaries/Plate_Boundaries.shp")
```

```python
print(boundries.head())
```

```python
boundries["coordinates"] = boundries.apply(lambda x: [(b,a) for (a,b) in list(x.geometry.coords)], axis = "columns")
boundries.drop('geometry',axis = 1,inplace = True)
print(boundries.head())
```

```python
japan_map = folium.Map(location = [35,136],titles= "cartodbpositron",zoom_start=5)

for i in range(len(boundries)):
    folium.PolyLine(locations = boundries.coordinates.iloc[i], width = 2, color = "Red").add_to(japan_map)
japan_map
```
