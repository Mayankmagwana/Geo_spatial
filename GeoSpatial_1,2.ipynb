You are a bird conservation expert and want to understand migration patterns of purple martins. In your research, you discover that these birds typically spend the summer breeding season in the eastern United States, and then migrate to South America for the winter. But since this bird is under threat of endangerment, you'd like to take a closer look at the locations that these birds are more likely to visit.



```python
bird = pd.read_csv("Kaggle/purple_martin.csv",parse_dates= ["timestamp"])
```

```python
bird["tag-local-identifier"].value_counts()
```

```python
bird.head()
```

```python
martin = gpd.GeoDataFrame(bird, geometry = gpd.points_from_xy(bird['location-long'] , bird['location-lat']))
martin.crs = {"init" : "epsg:4329"}
```

```python
type(martin)
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
america = world.loc[world["continent"].isin(["North America","South America"])]
america.plot()
```

```python
from shapely.geometry import LineString
```

```python
path_df = bird.groupby("tag-local-identifier")["geometry"].apply(list).apply(lambda x: LineString(x)).reset_index()

path = gpd.GeoDataFrame(path_df, geometry = path_df.geometry)
path.csr = {"init" : "epsg:4329"}
path.head()
```

```python
start_df = bird.groupby("tag-local-identifier")["geometry"].apply(list).apply(lambda x: x[0]).reset_index()

start = gpd.GeoDataFrame(start_df, geometry = start_df.geometry)
path.csr = {"init": "epsg4329"}
path.head()
```

```python
end_df = bird.groupby("tag-local-identifier")["geometry"].apply(list).apply(lambda x: x[1]).reset_index()

end = gpd.GeoDataFrame(end_df, geometry = end_df.geometry)
end.csr = {"init":"epsg:4329"}
end.head()
```

```python
ax = america.plot(figsize=(10,10), color = "none",edgecolor ="blue",linestyle =":")
start.plot(color = "orange",ax=ax)
end.plot(color = "green",ax=ax)
path.plot(color = 'black',linestyle = ':',ax=ax)
#ax.set_xlim([])
ax.set_ylim([-40,60])
```
